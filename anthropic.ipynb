```python
from langchain_anthropic import ChatAnthropic
from langchain_core.prompts import ChatPromptTemplate

chat = ChatAnthropic(temperature=0, model_name="claude-3-opus-20240229")

system = (
    "You are a helpful assistant that translates {input_language} to {output_language}."
)
human = "{text}"
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", human)])

chain = prompt | chat
response = chain.invoke(
    {
        "input_language": "English",
        "output_language": "Korean",
        "text": "I love Python",
    }
)

print(response)
```

```python
from langchain_anthropic import ChatAnthropic
from langchain_core.prompts import ChatPromptTemplate

chat = ChatAnthropic(temperature=0, model_name="claude-3-opus-20240229")

system = (
    "You are a helpful assistant that help people solve simple and complex math problems. Be precise and accurate in your response and don't hallucinate in your response."
)
human = "{text}"
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", human)])

chain = prompt | chat
response = chain.invoke(
    {
        "text": "Discuss the Barber Paradox in the context of self-reference."
    }
)

print(response)
```

```python
# streamlit app with claude 3
import streamlit as st
from langchain_anthropic import ChatAnthropic

st.title("Anthropic - Claude 3 Opus testing")

# Input for Anthropic API Key
anthropic_api_key = st.sidebar.text_input('Enter Anthropic API Key', type='password')

# Function to generate response
def generate_response(input_text):
    if not anthropic_api_key.startswith('sk-ant-'):
        st.warning('Please enter a valid Anthropic API key')
    elif not input_text:
        st.warning('Please enter some text')
    else:
        llm = ChatAnthropic(temperature=0, model_name="claude-3-opus-20240229")
        with st.spinner("Generating response..."):
            response = llm(input_text)
        st.info("Generated Response:")
        st.write(response)

# Streamlit form for user input
with st.form('my_form'):
    st.header("Enter Text")
    text_input_placeholder = "Enter your text here..."
    text = st.text_area("Input:", text_input_placeholder)
    submit_button = st.form_submit_button(label='Generate Response')

# Generate response when the form is submitted
if submit_button:
    # Use a default placeholder if the user didn't enter any text
    text_to_generate = text if text != text_input_placeholder else "What are the three pieces of advice for learning how to code?"
    generate_response(text_to_generate)
```
